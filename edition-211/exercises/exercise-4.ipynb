## A bit of philosophy


A lot has been said about the mystery of quantum. It is hard to understand because there is nothing to understand. There is the classical world, with classical bits which are True or False; and the quantum world with the Qubit that can have an infinity of states. But between the Classical and the Quantum worlds, there is veil that we cannot see through. We do not have knowledge of the quantum state of a qubit; what we know is the classical outcomes of measurements of the qubit. 

This is an absolute truth. We know nothing of the substance, we only know its reflection in this world. 

There are corollaries to this fact. 

- The first is that the outcomes possible for the measurement determine what we can know of the state, and therefore actually determine to the best anyone can know, what the state was. The measurement determines the state.
- Once we have an outcome for the measurement, what the quantum state "really" was no longer exists. No additional measurements can ever now happen inconsistent with the measurement just taken, hence the past as been determined for all time.
- And finally, once measured, the quantum state becomes the state indicated by the measurement. As all we can know about a state is its measurement, and the measurement has been made, then the state and the measurement must be the same.

A further corrolary is that we cannot duplicate a quantum state, called _cloning_. If we did we could measure the state repeatedly, perhaps with different measurements, and thereby learn more than is possible.

The veil between quantum and classical is like that between a dream and awakening. The dream exists as an unkowable that we learn about by recall on awakening. As as we recall the dream we resolve its inexplicable logic by replacing it with our waking expectations. The observation of the dream provides it with only these sort of permitted interpretations and the dream content changes irrevocably to become one of the offered recollections.

#### The photon polarization example

A photon has linear polarization at any angle of the 360 degrees of a circle. A polariod filter provides the simplest measurement of the photo, providing and yes or no answer to the measurement. A yes answer also means the photo passes through the filter; and a no answer means the photon does not pass. It must be absorbed and turned to heat.

It is also a remarkable measuring device because it can measure at any angle, simply by rotating the filter. You don't need different polaroids for different angles. It is a parameterized measurement instrument. Set it up at  and angle and it will pass and photon or not according to the measurement at that angle.

Once the measurement is taken, for the photons that pass, the polarization is exactly in the direction of the filter. What the polarization once was is of no consequence. It entered the filter, got measured, became the state consistent with the measurement, and then was absorbed or it passed through.

Experiments with these photos will show (similar to the Stern-Gerlach experiments (1922) on electron spin) that once a photon has a given polarization it will continue to  measure with certainty that measure. Photons once passed through the filter will competely pass through further filters that are  similarly rotated. However, they will be completely stopped if the follow on filter is at right angles in polarization. From this we learn that the measurement gives on  bit of information: whether the photon is at a certain angle, or at 90 degrees from that angle. 

Furthermore, if we knew that the photon was polarized, say either 0 or 90, then we would know by one measurement exactly which of the two cases held. A photon with 0 polarization will certainly be measured a 0 and subsequently be polarized at 0. A photon with 90 polarization will certainly be measures a 90 and, modifying our materials so that such photons pass as do the 0 degree photons, rather than be absorbed, subsequently it will be polarized at 90.

However, photons not polarized at 0 or 90 will by the laws of quantum be measured on this basis and we will learn very little about the true state before the measurement. The laws of quantum say that the more aligned with 0, the more likely the photon will measure 0, but it can measure 90; and a photon measured at 45 will randomly with 50/50 probability be measured at 0 and 90.

If we had a stream of photons, and we were told they were all at angle x, then by repeated measurement we can statistically learn x; however, since we cannot clone a photon, there must be some known mechanism that assures us they are all at angle x. We are learning about the decisions of the mechanism, in a sense, which  is classical knowledge, as it was constructed in a classical way, even if the techniques and materials were quantum.


#### The electron energy levels example

The photon polarization example does not discover the full mystery of quantum, as we still can believe that the photon, ahead of its measurement, was polarized at a particular angle. So that the limitation of being able to only measure on a pair of orthogonal directions, with an answer of which of the two directions is the result of the measurement as well as the current polarization of the photon, seems to be something that perhaps we can fix.

Perhaps there is a measurement apparatus with a dial that reads analog zero to 360, and the fault is with the measuring device. That there is nothing fundamental here.

An electron can be in several energy levels. For a qubit system, we consider that it can be in one of two energy levels. It cannot be in a fractional level. It is a _quantum leap_, a jump of one quantum of energy, to go from one level to another. 

However, which energy level it is in  is a quantum phenomena, and when we measure it becomes classical. 

If we knew that the electron was in one or the other levels, the measurement would perfectly measure that. But an electron can be in a _superposition_ of levels. It can be, in analogy with the photon polarization example, "at 45 degrees". It can be simultaneously in both levels.

What this means is that the measurement of the electron level will half the time show it in one level, and it will now be in that level, and half the time in the other level. 

You should now understand that the original photon example was misleading. Although all of what we see in the classical world leads us to believe in 45 degree polarization of a photon, all we ever can know is the result of a measure along a single orthongonal basis. And this is exactly all we can know about this case, of electron levels.

In fact, the 45 degree polarization is conceptually equal to a photon in a superposition of 0 and 90. It all depends on how you measure it. 




## Observables


A _measurement_ is the application of an observable. An _observable_ is an operator that allows certain _outcomes_, which are real numbers, and associates a preferred state, the _basis state_ to each outcome.

Applying a observable to a quantum state announces to the classical an outcome, and places the quantum state in the basis state associated with that outcome. 

So far we have experienced two observables, $X$ and $Z$. These are also operators, but can also be obervables through the following decompositions,

\begin{eqnarray}
Z &=&  |0\rangle\langle 0| - |1\rangle\langle 1| \\
X &=&  |+\rangle\langle +| - |-\rangle\langle -| 
\end{eqnarray}

The symbols such as $|0\rangle\langle 0|$ are _projection operators_ which measure a state in the direction of a given qubit, in this case the qubit $|0\rangle$, and create an output qubit of that size in that direction. What we need to know right now is that, 


$$
\bigl( |a\rangle\langle b|\bigr)\,|c\rangle =  |a\rangle\,\bigl( \langle b|c\rangle\bigr)
  = |a\rangle\langle b|c\rangle
$$

where the last group of symbols drops all parenthesis since the order of parenthesis does not matter, and that the symbol $\langle a|b\rangle$ is 1 for equal qubits $a$ and $b$, and zero for opposive qubits. The opposite qubits we already know are $|0\rangle$ and $|1\rangle$ are opposites, as well as $|+\rangle$ and $|-\rangle$ are opposites.

And therefore, 

\begin{eqnarray}
Z\,|0\rangle &=&   \bigl(|0\rangle\langle 0| - |1\rangle\langle 1| \bigr)\,|0\rangle \\
&=&  \bigl(|0\rangle\langle 0|\bigr)\,|0\rangle - \bigl(|1\rangle\langle 1| \bigr)\,|0\rangle \\
&=&  |0\rangle\,\bigl(\langle 0|0\rangle\bigr) - |1\rangle\,\bigl(\langle 1|0\rangle  \bigr)\\
&=&  |0\rangle\, 1 - |1\rangle\,0\\
&=& |0\rangle \\
& & \\
Z\,|1\rangle &=&   \bigl(|0\rangle\langle 0| - |1\rangle\langle 1| \bigr)\,|1\rangle \\
&=& \ldots \\
&=& -|1\rangle
\end{eqnarray}

We will use linear algebra as a method of computation, but if we were to take linear algebra seriously, we could define operator as a Unitary linear operator, and a measurable as an Hermitian linear operator.

```python
import math
import random
import numpy as np

np.set_printoptions(precision=2,floatmode='fixed',suppress=True)

"""
A qubit is a vector of two complex numbers with length one. Restricting to real numbers, this
means a qubit is a point on the unit circle. The four points we have explored are geometrically
angles 0, 90, 45 and -45. Here they are.
"""

def qubit_zero():
    """
    returns a |0>, as a linear algebra object
    """
    return np.array([1.0,0.0])

def qubit_one():
    """
    returns a |1>, as a linear algebra object
    """
    return np.array([0.0,1.0])

def qubit_plus():
    """
    returns a |+>, as a linear algebra object
    """
    q_plus = 1.0/math.sqrt(2.0) * ( qubit_zero() + qubit_one() )
    return q_plus

def qubit_minus():
    """
    returns a |->, as a linear algebra object
    """
    q_plus = 1.0/math.sqrt(2.0) * ( qubit_zero() - qubit_one() )
    return q_plus

def not_a_qubit():
    """
    for whatever reason, returns the zero vector, which is not a qubit because it 
    does not have length 1
    """
    return np.zeros(2)


"""
We also have several operators, X, Z and H are the most used so far. Here they are:
"""

def operator_X():
    """
    returns the X operator, as a linear algebra object
    """
    return np.array([[0.0,1.0],[1.0,0.0]])

def operator_Z():
    """
    returns the Z operator, as a linear algebra object
    """
    return np.array([[1.0,0.0],[0.0,-1.0]])


"""
We had some functions described about. here they are in linear algebra code.
"""

def bracket_operator_real(bra,ket):
    """
    from <bra| and |ket> return <bra|ket> only correct for real bra and ket
    """
    return np.inner(bra,ket)

def projection_operator(ket):
    """
    from ket, return |ket><ket|
    Note: this uses np.conj to make a correct answer for complex ket's. 
    Compare to bracket_operator_real for the difference.
    """
    return np.outer(ket,np.conj(ket))

def apply_operator(operator, ket):
    """
    from |a><a| and |b> return |a><a|b>
    """
    return np.dot(operator,ket)
    

print(f"\nwe know the qubit states as symbols. here is how we make them vectors\n\tsymbol\t->\tvector\n\t|0>\t->\t{qubit_zero()},\n\t|1>\t->\t{qubit_one()}," 
      + f"\n\t|+>\t->\t{qubit_plus()}\n\t|->\t->\t{qubit_minus()}")

print(f"\nwe know the observations as X and Z. here is how we make them matrices\n\tsymbol\t->\tmatrix\n\tX\t->\t{operator_X()[0]}\n\t\t\t{operator_X()[1]}"
          +f"\n\tZ\t->\t{operator_Z()[0]}\n\t\t\t{operator_Z()[1]}" 
     )

```

## Calculating the probabilities

We have seen two observables $X$ and $Z$ that were also quantum gates. We can put a state through the gate, example $Z\,|0\rangle$, but we want to measure the state by the observable. We have the observable writen separating out the two outcomes,

$$
M = e_1\, |m_1\rangle\langle m_1| + e_2\, |m_2\rangle\langle m_2| 
$$

Where $m_i$ are the measurement basis, and $e_2$ are the outcomes. The probability $p_i$ of measuring outcome $i$ is on state $|\phi\rangle$ is,

$$
p_i = \langle\phi|\bigl(|m_i\rangle\langle m_i|\bigr)|\phi\rangle = \langle\phi|m_i\rangle\langle m_i|\phi\rangle
$$

While the symbol $\langle \phi |$ has not be introduced, it's called a bra, and the combining with a ket, makes a bracket, such as $\langle m_i|\phi\rangle$. You already know how the bracket works on pairs of elements in the basis. Some more things you need know about the bracket. 
- It is linear in the second place, $\langle \psi|\,\,\alpha\phi_1+\beta\phi_2\,\,\rangle = \alpha \,\langle\psi|\phi_1\rangle + \beta\,\langle \psi|\phi_2\rangle $.
- It has a twisted symmetry, $\langle \psi | \phi\rangle = \langle \phi | \psi\rangle^*$, where the star is complex conjugation.

And with this we can evaluate the bracket on a general state $\phi = \alpha \,|0\rangle + \beta\,|1\rangle$,

$$
\langle m_i|\,\,\alpha |0\rangle + \beta |1\rangle\,\,\rangle = \langle m_i| \,\,\alpha |0\rangle\,\,\rangle + \langle m_i|\,\,\beta |1\rangle\,\,\rangle = \alpha\,\langle m_i |0\rangle + \beta\,\langle m_i|1\rangle = 
\begin{cases}
\alpha &\mbox{if } i=0 \\
\beta &\mbox{if } i=1
\end{cases}
$$

If $\alpha$ and $\beta$ are real, then the twisted symmetry has no complex numbers to conjugate, so,

$$
p_i = \langle\phi|m_i\rangle\,\langle m_i|\phi\rangle = \langle m_i|\phi\rangle^2 = 
\begin{cases}
\alpha^2 &\mbox{if } i=0 \\
\beta^2 &\mbox{if }i=1
\end{cases}
$$

And if the outcome was $i$, then the new state is,

$$
\frac{
|m_i\rangle\langle m_i|\phi\rangle
}{\sqrt{p_i}}
$$

not that in the real case, $\langle m_i|\phi\rangle$ is already _a_ square root of $p_i$, so the 
formula basically decides between $|m_i\rangle$ and $-|m_i\rangle$.

In the case of complex numbers, we have some slight changes to make. All of which adds to the interest.
